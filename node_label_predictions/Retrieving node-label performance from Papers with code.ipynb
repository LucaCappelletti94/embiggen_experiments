```python
k = 7
top_k = {
    paper
    for paper, _ in sorted(papers_citations.items(), key=lambda x: x[1], reverse=True)[:k]
}
```

```python
top_k
```

```python
tables = []
for graph, html_path in pages.items():
    table = pd.read_html(html_path)[0]
    table = table[table.PAPER.isin(top_k)]
    table = table.groupby(["MODEL", "PAPER"]).first()
    table = table[table.ACCURACY.notna()]
    table = table[table["TRAINING SPLIT"].isna()].dropna(axis=1)
    table["Accuracy"] = table["ACCURACY"].str.strip("%").map(lambda x: float(x)/100)
    table["graph"] = graph
    table = table.drop(columns=["RANK", "YEAR", "ACCURACY"])
    tables.append(table)
pd.concat(tables).dropna()
```
